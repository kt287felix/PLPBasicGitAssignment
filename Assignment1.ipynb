### Encoding Categorical Variables

```python
# Categorical columns
cat_col = [col for col in df.columns if df[col].dtype == 'object']
print('Categorical columns :',cat_col)
# Numerical columns
num_col = [col for col in df.columns if df[col].dtype != 'object']
print('Numerical columns :',num_col)
```

```python
df[cat_col].nunique()
```

```python
# Booleans dummies
df['Number of Cities Lived'] = df['Number of Cities Lived'].astype(str)
df['Close Relative'] = df['Close Relative'].astype(str)
```

```python
pd.get_dummies(df['Close Relative']).astype(int).head()
```

#### Dropping 'Customer ID'

```python
df = df.drop(['Customer ID'], axis=1)
```

```python
df = pd.get_dummies(df).astype(int)
df.head()
```

### Exploratory Data Analysis

```python
df.describe()
```

### Feature Selection

```python
x = df.drop('Avg Bill', axis=1)
y = df['Avg Bill']
```

```python
# Checking correlation
corr = x.join(y).corr()
corr
```

### Splitting the Data

```python
from sklearn.model_selection import train_test_split


x_train, x_test , y_train , y_test = train_test_split(x,y,test_size = 0.3,random_state=42)
train_data = x_train.join(y_train)
```

### Train,Building and Evaluate Models

#### 1.Linear Regression


```python
# model for linear regression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

model_regress = LinearRegression()
model_regress.fit(x_train,y_train)

# make prediction for the model
y_pred_model_regress = model_regress.predict(x_test)

#Evaluate the model using measures of accuracy in regression
mae_model_regress = mean_absolute_error(y_test,y_pred_model_regress)
mse_model_regress = mean_squared_error(y_test, y_pred_model_regress)
r2_model_regress = r2_score(y_test, y_pred_model_regress)

print(f'Linear Regression - MAE: {mae_model_regress}, MSE: {mse_model_regress}, R2: {r2_model_regress}')
```

### 2.Decision Tree Regressor

```python
# Train the model
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)
dt.fit(x_train, y_train)

# Make predictions
y_pred_dt = dt.predict(x_test)

# Evaluate the model
mae_dt = mean_absolute_error(y_test, y_pred_dt)
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

print(f'Decision Tree Regressor - MAE: {mae_dt}, MSE: {mse_dt}, R2: {r2_dt}')
```

### 3.RandomEorest Regressor

```python
# Train the model
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(x_train, y_train)

# Make predictions
y_pred_rf = rf.predict(x_test)

# Evaluate the model
mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print(f'Random Forest Regressor - MAE: {mae_rf}, MSE: {mse_rf}, R2: {r2_rf}')
```

### 4.Gradient Boosting Regressor

```python
# Train the model
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(random_state=42)
gbr.fit(x_train, y_train)

# Make predictions
y_pred_gbr = gbr.predict(x_test)

# Evaluate the model
mae_gbr = mean_absolute_error(y_test, y_pred_gbr)
mse_gbr = mean_squared_error(y_test, y_pred_gbr)
r2_gbr = r2_score(y_test, y_pred_gbr)

print(f'Gradient Boosting Regressor - MAE: {mae_gbr}, MSE: {mse_gbr}, R2: {r2_gbr}')
```

### Support Vector Regressor(SVR)

```python
# Train the model
from sklearn.svm import SVR

svr = SVR()
svr.fit(x_train, y_train)

# Make predictions
y_pred_svr = svr.predict(x_test)

# Evaluate the model
mae_svr = mean_absolute_error(y_test, y_pred_svr)
mse_svr = mean_squared_error(y_test, y_pred_svr)
r2_svr = r2_score(y_test, y_pred_svr)

print(f'Support Vector Regressor - MAE: {mae_svr}, MSE: {mse_svr}, R2: {r2_svr}')
```

### Model Perfprmance

```python
# Create a DataFrame to compare the results
results = pd.DataFrame({
    'Model': ['Linear Regression', 'Decision Tree', 'Random Forest', 'Gradient Boosting', 'SVR'],
    'MAE': [mae_model_regress, mae_dt, mae_rf, mae_gbr, mae_svr],
    'MSE': [mse_model_regress, mse_dt, mse_rf, mse_gbr, mse_svr],
    'R2': [r2_model_regress, r2_dt, r2_rf, r2_gbr, r2_svr]
})

print(results)
```
